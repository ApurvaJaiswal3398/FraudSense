```python
import pandas as pd
```

```python
df=pd.read_csv(r'dataset\DATASET.csv')
```

```python
cash_in_df = df[df['type'] == 'CASH_IN'][df['newbalanceOrig'] == df['oldbalanceOrg'] + df['amount']]
cash_in_df
```

```python
cash_in_df['nameDest']=0
cash_in_df['oldbalanceDest']=0
cash_in_df['newbalanceDest']=0
CASH_IN=cash_in_df
```

```python
CASH_IN
```

```python
CASH_IN.to_csv('CASH_IN_DETAILS.csv')
```

```python
cash_out_df = df[df['type'] == 'CASH_OUT'][df['newbalanceOrig'] == df['oldbalanceOrg'] - df['amount']]
cash_out_df
```

```python
cash_out_df['nameDest']='N/A'
cash_out_df['oldbalanceDest']=0
cash_out_df['newbalanceDest']=0
CASH_OUT=cash_out_df
```

```python
CASH_OUT
```

```python
CASH_OUT.to_csv('CASH_OUT_DETAILS.csv')
```

```python
trans_df = df[df['type'] == 'TRANSFER'][df['newbalanceOrig'] == df['oldbalanceOrg'] - df['amount']]
trans_df
```

```python
TRANS = trans_df[trans_df['newbalanceDest'] == trans_df['oldbalanceDest'] + trans_df['amount']]
TRANS
```

```python
TRANS.to_csv('TRANSFER_DETAILS.csv')
```

```python
debit_df = df[df['type'] == 'DEBIT'][df['newbalanceOrig'] == df['oldbalanceOrg'] - df['amount']]
debit_df
```

```python
DEBIT = debit_df[debit_df['newbalanceDest'] == debit_df['oldbalanceDest'] + debit_df['amount']]
DEBIT
```

```python
DEBIT.to_csv('DEBIT_DETAILS.csv')
```

```python
pay_df = df[df['type'] == 'PAYMENT'][df['newbalanceOrig'] == df['oldbalanceOrg'] - df['amount']]
pay_df
```

```python
pay_df['newbalanceDest']=pay_df['oldbalanceDest']+pay_df['amount']
PAYMENT=pay_df
```

```python
PAYMENT
```

```python
PAYMENT.to_csv('PAYMENT_DETAILS.csv')
```

```python
FRAUD=pd.concat([CASH_IN,CASH_OUT,TRANS,PAYMENT,DEBIT]).drop_duplicates(keep=False)
```

```python
FRAUD = FRAUD.drop(['Unnamed: 0'], axis=1)  # drop the first column
```

```python
FRAUD.to_csv('FRAUD_DETECTION.csv')
```
